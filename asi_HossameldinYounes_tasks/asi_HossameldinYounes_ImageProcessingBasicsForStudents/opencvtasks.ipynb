```python
import cv2
import numpy as np

img = np.zeros((600,600,3),np.uint8)

delete = False

def draw_circle(event,x,y,flags,parameter):  
    global delete
    if event == cv2.EVENT_LBUTTONDOWN:
        circles.append([x,y])
    elif event == cv2.EVENT_RBUTTONDBLCLK:
        circles.clear ()
        delete = True
    elif event == cv2.EVENT_RBUTTONDOWN:
        circle = circles[-1]
        circle = (circle[0],circle[1],'delete')
        circles[-1] = circle


circles = []

while True:
    
    cv2.imshow('MY IMG',img)
    cv2.setMouseCallback('MY IMG',draw_circle)

    for circle in circles:
        if len(circle) == 3:
            cv2.circle(img,(circle[0],circle[1]),20,(0,0,0),2)
            circles.pop()
        else:
            cv2.circle(img,(circle[0],circle[1]),20,(255,0,0),2)


    if delete:
        img = np.zeros((600,600,3),np.uint8)
        delete = False
    
    k = cv2.waitKey(10)
    if k == ord('q'):
        break

cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np

counter = 0
array = np.zeros((4,2),int)
def addpts(event,x,y,flags,parameter):
    global counter,array
    if event == cv2.EVENT_LBUTTONDOWN:
        array[counter] = x,y
        counter = counter + 1


width,height = 280,350
img = cv2.imread("resources/computer vision.PNG")

cv2.namedWindow(winname='MY_IMG')
cv2.setMouseCallback('MY_IMG',addpts)



while True:
    if counter == 4:
        pts1 = np.float32([array[0],array[1],array[2],array[3]])
        pts2 = np.float32([[0,0],[width,0],[0,height],[width,height]])
        matrix = cv2.getPerspectiveTransform(pts1,pts2)
        output = cv2.warpPerspective(img,matrix,(width,height))
        cv2.imshow('output IMG',output)
    cv2.imshow('MY_IMG',img)
    cv2.waitKey(1000)
```
